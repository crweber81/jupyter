## Properties of Estimators

As King (1998), we're primarily concerned with making inferences about a population from a sample. Recall, in the frequentist tradition, we think of parameters as fixed characteristics of the population; statistics are derived from sample(s) drawn from the population. They are often qualified with some degree of uncertainty. An estimator is the formula or equation used to represent what we think is the process governing a parameter. So, $y_i=\alpha+\beta x + \epsilon$ forms the relationship between x and y in the population, governed by fixed parameters $\alpha$, $\beta$, and an error term. This is not the estimator. The estimator is the method we then use to guess these parameters. Ordinary least squares is an estimator. It's one of many estimators. Much of POL 681 was oriented around showing it is the best linear unbiased estimator, in particular conditions.

The issue is that we only have a sample, or set of samples. Thus we wish to draw an inference about the population from the sample. The various properties of an estimator can be explored in several ways. One is to assume a sample is observed over repeated trials, so we have repeated samples. We could then perform our estimation procedure across samples and compile these estimates. Recall the logic of the central limit theorem, for instance.

If we were able to repeatedly draw samples and run our estimation procedure, we are left with a \textit{sampling distribution}. We can then explore properties of an estimator by examining this distribution. Most notably, the distribution will have mean 

$$E(\hat {\theta})$$ 

$$var(\hat{\theta})=E(\hat{\theta}-E(\hat{\theta}))^2$$

With $\sqrt{var(\hat{\theta})}$ as the standard deviation of the sampling distribution. We can then establish three statistics to explore the properties of an estimator. The ***sampling error*** is the deviation between our estimator $\hat{\theta}$ and the population parameter ($\theta$). There will always be some degree of error by drawing a sample from a population. We may wish to minimize this sampling error, but we shouldn't expect it to be zero. 

***Bias*** is the difference between the expected value of an esimator relative to the true population parameter, i.e., $\hat{\theta}-\theta$. Note how this is different from the sampling error. The sampling error refers to the difference for a single estimate produced from our estimator. Bias is the average of many samples analyzed with an estimator.

We're often also interested in how much an estimator varies about a true population parameter. Here, let's define the ***Mean Squared Error*** (MSE), or $E(\hat{\theta}-\theta)^2$. Again, notice the difference here. Here, we are concerned with how much an estimator varies around the true population parameter. It can be shown that the MSE may be rewritten as

$$E(\hat{\theta}-E(\hat{\theta}))^2 +[E(\hat{\theta}-\theta)]^2$$ 

In other words, the MSE is a function of the variance of the estimator and the bias of the estimator. We generally wish for this value to be small. For instance, if we were to compare two estimators, we should prefer the estimator that is unbiased with minimum variance, which of course translates to having the smallest MSE.

#### Finite Sample Properties

Much of what you've learned in POL 681 pertained to OLS and **small sample properties**. Some estimators have desirable properties even when a sample is small. For instance, OLS has desirable small sample properties, which really only means the average of many samples(?).

***Unbiasedness***, or $E(\hat{\theta})=\theta$. Over repeated samples, the mean of the sampling distribution will equal the true population parameter. Although it may seem that unbiaseness is sufficient to draw conclusions about an estimator, it's not. The reason is that it refers to repeated samples. We have no idea if a parameter estimate from a single sample is near to the true population value.

***Efficiency***, or $var(\hat{\theta})<var(\tilde{\theta})$. Compared to other estimators, our estimator should have smaller variance. 

In some cases, there is a tradeoff, in that we may have an estimator that is known to be biased, but it has smaller variance. Or, we have an estimator that is unbiased, but has huge variance. In fact, we can operationalize this further by simply using the MSE. We may prefer an estimator that has the smallest MSE, even if it's known to be biased. 

An example of this is ridge regression. This is the case with ridge regression. We'll talk about this later in the semester. The ridge estimator will have a smaller MSE in some cases.

Something that we haven't sufficiently established yet is the **large** or **asymptotic properties** of an estimator. What happens to the behavior of the sampling distribution as the sample size approaches infinity. 

***Asymptotic Properties***.
These refer to the properties of an estimator as the sample size approaches infinity. In other words, as the sample size gets larger and larger, what happens to the characteristics of an estimator? An estimator is said to be \emph{asymptotically unbiased} if 

$$\lim_{n\to\infty} E(\hat{\theta})=\theta$$

A perfect example of this is the estimate of the sample variance; you may recall that the population variance is: 

$$\sigma^2={{1}\over{n}}(x-\mu_{x})^2$$

But the sample variance is:

$$\sigma^2={{1}\over{n-1}}(x-\bar{x})^2$$

But, the population variance is an unbiased estimator as the sample size increases. In particular,

$$\lim_{n\to\infty} E(\hat{\sigma^2})=\lim_{n\to\infty} [{{n-1}\over{n}}]\sigma^2=\sigma^2$$

Note that as n approaches infinity, the term $[{{n-1}\over{n}}]$ approaches 1 and the estimator converges to the population parameter. 

An estimator is ***consistent*** if:

$$P\lim_{n\to\infty}(|\hat{\theta}-\theta|<d)=1$$

then,

$$Plim_{n\to\infty}(\hat{\theta})=\theta$$

Really, this is subtly different from asymptotic unbiasedness. What it means is that the probability that the estimator is not different from the population parameter is 1 as the sample size approaches infinity. Put slightly different, the probability increases to 1 that as the sample size increases the estimator converges to the population parameter. It is possible -- in fact quite common -- for an estimator to be biased but consistent. We sometimes refer to this as the bias/variance tradeoff. A prime example is the variance estimate above. When we divide by $n$ this is also called the MLE estimatate of the variance. It is biased in small samples, but consistent and asymptotically unbiased. I like to think of consistency more in terms of the variance of an estimator, such that $P\lim_{n\to\infty}$ represents the point at which the variance is 0 and the entire distribution collapses to one value -- i.e., $\lim_{n\to\infty} MSE(\hat{\theta}=0)$

$$RSS=\sum_{i=1}^N=Y_i-(\beta_0+\beta_1 x_{1,i}+\beta_2 x_{2,i}+\dots+\beta_k x_{k,i})$$

### Bayesian Inference

Notice that the maximum of the **log** of the likelihood function is the same as the same as the maximum of the likelihood function. Each observation contributes something to the log likelihood. The "frequentist" label may make a bit more intuitive sense: Assume each observation is the realization of some draw from a probability distribution with fixed parameters. We're finding The parameter value(s) that maximizes the probabiity of observing the data. The language is clunky, likely for the same reason that the frequentist interpretation of probability is  counterintuitive. We are estimating a value of $\theta$ that simply maximizes the probability of 1,0,1,0. Here it's 0.5. We're fixing $\theta$, we can't say anything about the distribution of $\theta$. It's simply some value or set of values that maximum the likelihood of observing a particular dataset. It is $p(Data|\theta)$. But what about  $p(\theta|Data)$. Using Bayes' rule perhaps we have some prior knowledge about $\theta$.

### How is Bayesian Inference Different?

Frequentism relies on a compelling, though narrow application of probability. What do we observe in repeated trials? This model is hard to apply to the social sciences, when we only observe a particular outcome once. While Bayesian and Frequentists may settle on similar probability estimates, the overall interpretation is quite different. 

#### Frequentist and Bayesian Inference

Generate an estimate of $\theta$ such that the joint probability of observing the data, given theta, maximizes the probability of observing the data. In the case of ML, we specify the **data generating function**, where the PDF is $p(y)=\prod^N_{i=1}\theta^{y_i}(1-\theta)^{(1-y_i)}$. Again, it's no more complicated than finding an optimal value of $\theta$, or the most likely value of $\theta$ that produces a particular data set. It is finding the maximum of $p(Data|\theta)$. Remember, Bayes' Rule is:

$p(\theta|Data)\propto p(Data|\theta) p(\theta)$

The **posterior** distribution of $\theta$ is proportional to the ***likelihood*** times the ***prior***

Recall, we can invert this statement through Bayes' rule, as $p(\theta|Data)$. This is more than changing the conditional probability -- we are now treating the parameters as stochastic variables, and the data are fixed. Note, this is a more satisfying way to view statistical models. A 95\% confidence interval is not the expectation about what happens in repeated draws. Instead, it is based from the posterior distribution of the parameter estimate, which follows a particular probability distribution.

Let's think about this with a few examples. Maximum likelihood is an optimization tool -- it helps us find the best mapping of $\theta$ applied to an observed dataset. It's the application of maximum likelihood to frequentist probability that is strange. Say 538 predicts a 72/100 chance that Biden wins the election. Of course, the prediction is based on a whole host of important and hard to interpret assumptions. The 0.72 estimate means if the election were held now, and then again under equivalent circumstances over and over again, Biden would win 72 times out of 100. It's a probability statement that only applies to the present. Frequentism is quite reasonable in games of chance, it's also applicable in scientific experiments where lab conditions stay constant. It's harder when applied to observational social data. Probability in these models only applies to the present. It's sort of a Groundhog's Day interpretation of chance. Model parameters are fixed. Another thing to remember is that inference stems from repeated draws from a population. Under frequentism one shouldn't assume the model parameters are random. They are fixed. Uncertainty is built into sampling.

Bayesian inference flips things around. Now we view the data as fixed, the parameters are random. This is quite important for quantitative social research. Inference is not based in uncertainty in the data. Uncertainty stems from uncertainty about the model parameters. After all, the data are fixed. The model parameters do not depend on the data. We can apply the posterior parameter estimates to any data -- representative or not -- and make valid inferences about future events. For these reasons, Bayesian inference is often seen as more suitable to future prediction; it's a more intuitive way to extrapolate one's model to future events.





This is really useful -- if we are able to formulate expectations about a model parameter, we can incorporate that into our prior. A diffuse prior means the mode of the posterior will be the same as the likelihood solution. Try a more informative prior.

At this point, you should be familiar with the Bernoulli and Binomial densities. The Binomial density is used if we observe a series of Bernoulli trials. Thus, the Bernoulli is a special case of the Binomial, i.e., when $n=1$. We use the Bernoulli/Binomial when the observation for each unit is "success" or "failure," or equivalently, 0 or 1. In other words, we have a binary variable. The Bernoulli density for $n$ observed trials is simply the product of each constituent probability. 

$$p(k)=\prod\theta^y_i(1-\theta)^{1-y_i}=\theta^k(1-\theta)^{n-k}$$

We can calculate the probability of any set of $k$ outcomes, given $n$, by multiplying the above function by the number of $k$ possible combinations for $n$ trials, i.e., $n \choose k$

$$p(K | \theta, N)= {n \choose k} \theta^k(1-\theta)^n-k$$

Again, it's more analytically and computationally tractable to apply the logarithm to the function. We can just take the product of Bernoulli trials across all draws -- it's the same thing.
  
$$  log(p(D|\theta))=\sum_{n=1}^Nlogp(y_n|\theta)=\sum_{n=1}^N[y_n log\theta+(1-y_n)log(1-\theta)] $$
The key ``unknown'' parameter in these trials is $\theta$, the probability of success ($p[y=1]$). We could use either the Bernoulli or Binomial densities (I'll use the two interchangably), but instead of estimating $\theta$ directly, we write $\theta$ as some function of predictor variables. 

The simplest thing to do would be to assume that $\theta_i=a+bx+\epsilon$. Let's assume theta is the probability of voting for Donald Trump. In this case, maybe we assume $x$ is conservatism.

The problem, however, is that we never actually observe $\theta$, like we do $y$ in a linear regression; instead, we only observe an outcome that is 0 or 1, which of course is not a probability.

This can be viewed as a problem of missing data. Assume $y_{obs} \in [0,1]$. Likewise, let's assume that $x$ is just some n-length vector of real numbers, $x \in \rm I\!R^{n}$. We want to estimate something like this, $(a+b_x+\epsilon) \rightarrow \theta \rightarrow y$. We need to map the linear predictor(s) onto a probability which is then used to estimate the joint probability of $n$ Bernoulli trials. The problem is that we do not have $\theta_i$. It's not an observed variable like $y$ when we ran a linear regression. It's an unkown parameter. The other restriction is that $\theta_i$ cannot be any number. By construction, it must be between 0 and 1. Let's start out with the simplest way to tackle this question.

***The Linear Probability Model***

It may seem fine to estimate a linear regression, regressing $y_{obs}$ on a set of covariates. Indeed, we could think of the outcomes as probabilities, where $E(y|x)=b_0+b_1x$, with expectation corresponding to probability estimate, $P_i$. This initially may appear okay because if we meaningfully scale y to be a 0 or 1, then we can interpret the any slope (or intercept) as an effect on the probability of observing 1. This is called the ***linear probability model***.

Without reading ahead, if $y \in [0,1]$, what is the problem with minimizing the sum of squared residuals associated with the difference of $\hat{y}$ and $y_{obs}$? Is there a problem?

There are several issues that arise when we use OLS with a binary dependent variable. First, the expectation is not a true probability because it is not bound between 0 and 1. There is nothing that restricts $0<\hat{y}<1$. It is entirely possible to have predictions outside [0,1]. This makes little sense in the context of a probability.  Put another way, we can very easily predict $y$ well outside the plausible observed values of 0/1. 

That's one strike against the model. Second, the residuals are not normally distributed, nor are they constant. Why? 
  
Recall that $e_i=y_i-\hat{y_i}$. But, we know that $y_i$ may only take on two values, 0 or 1. Thus, if $y=1$, the residual becomes: $1-b_0-b_1x$, but if $y=0$, the residual becomes $-b_0-b_1x$. We will always have a different variance estimate when $y=0$ relative to when $y=1$. Put another way, as noted in Long (1997, p. 38):

$$var(y|x)=pr(y=1|x)(1-pr(y=1|x))=xb(1-xb)$$

This is the second strike. We also encounter heteroskedasticity.

Now, we know that if the variance is not constant, there are several solutions. Insofar as $b$ is a non-zero number, we have heteroskedasticity. But, knowing the nature of heteroskedasticity, we could first estimate a model using OLS. Then, we could use generalized least squares assuming $var(e)=\hat{y}(1-\hat{y})$. This will partially solve the heteroskedasticity problem, though it will not solve the non-sensical prediction problem. 

Equally serious is the issue of functional form. The linear model assumes a constant effect of $x\rightarrow y$. It doesn't matter what value of $x$ we consider, the marginal effect on $y$ is always $b$. What if we expect a decreasing or increasing return of $x$ on $y$, such that the rate of change differs at levels of $x$? If OLS assumes the partial derivative with respect to $x$ is $b$, what if this is not the case?

Suppose we anticipated the following relationship between age and the probability of visiting a doctor on a regular basis.

We would incorrectly capture this relationship in a linear expression in which $age$ has a constant increasing effect on visiting a doctor. 
  
Thus, it seems like we need a different solution. Let's write down what we know. First, our data $y_{obs}$ are coded 0 or 1. We know that because of this, we might specify a the data follow a series of Bernoulli trials, where $Y_i \sim bern(\theta)$. 

We already know how to find the maximum likelihood estimate of $\theta$ from our observed data. The joint probability of our data, $p(y_1, y_2,..y_n)=\prod (p_i)$. If we are willing to assume that the product of these probabilities is guided by $\theta$ we can specify the Binomial. The probability of $Y_i=1$ is $\theta$; otherwise $1-\theta$. 

But, here's what we don't know: How do we specify that $\theta$ is (1) no longer a constant, and (2) a function of some covariates? We need to be sure that our predictions are bound between 0 and 1, and the probability $y=1$ may not necessarily be constant. This is essentially the motivating basis for logit and probit regression.

### Latent Variable Regression

All we have access to is a binary $y_{obs}$ and any set of covariates we collect. Now, underlying $y_{obs}$ let's assume everyone has a continuous $y_{latent}$ value.\footnote{Soon we will consider how this $y_{latent}$ variable maps onto $\theta$} This $y_{latent}$ value somehow maps onto $y_{obs}$. Also, let's just assume that $y_{latent} \in ({-\infty, \infty})$. It can take on any value on the number line. We can also assume that $y_{latent}$ is a function of some observed covariates, plus some error term, i.e, $y_{latent}=xB+e$. 

This is important: We have specified a linear equation with respect to this latent $y$. Of course, if we could observe this continuous variable, then we could use least squares. But, we cannot directly observe this variable. This explanatory model is often called the ***structural model***.

But we only observe $y_{obs}$, not $y_{latent}$. Thus, we need something else, which is called the ***measurement model***. This is the mapping of $y_{latent}$ onto $y_{obs}$. Let's assume there is some threshold that leads to $y_{obs}$.

$$y_{obs} = {1, y_{latent}>\tau}\atop {0,  y_{latent}\leq \tau}.$$

If an individual's score on the latent variable is greater than some threshold, $\tau$, this translates to a $y_{obs}=1$; otherwise, it is 0. So for instance, assume $\tau=0$ (It doesn need to be fixed; it's not identified with an estimated intercept. 0 is convenient).


If $y_{latent}>\tau$ -- i.e., any value above the the horizonal line -- then we observe $y=1$; otherwise, we observe $y=0$. 

Because the independent variable presumably does not perfectly predict $y_{latent}$, then $y_{latent}=x_iB+e_i$. Again, we assume error, or something left over after accounting for the linear effect of $x$ on $y_{latent}$. 

The problem is that we don't observe $y_{latent}$; nor can we observe $e_i$. It's not possible to observe the difference between our prediction of $y_{latent}$ and the (actual?) latent y. I realize the previous sentence makes little sense. It should make little sense. We are calculating the error for something that isn't actually observed. We need to make an assumption about the distribution of $e_i$. 

Traditionally, one makes an assumption that the errors are standard normal (mean$=$0, variance$=$1), which is the  ***probit regression***. This means the errors have mean 0 and standard deviation 1. Or, we assume that the errors follow the logistic distribution, which is the ***logit regression***. Logistically distributed errors have mean 0 and variance $\pi^2/3$. 

If the errors follow the standard normal, the PDF/CDF for the errors is:

### PDF for Normal Errors (Probit):
 
$$probit={{1}\over{\sqrt{2 \pi}}}exp({-{t^2}\over {2}})$$.

### CDF for Normal Errors (Probit):

$$probit=\int_{-\infty}^{e}{{1}\over{\sqrt{2 \pi}}}exp({-{t^2}\over {2}})dt$$. 

If the errors follow the logit density, then: 

### PDF for Logit Errors (Logit):
 
$$logit(e)={{exp(e)}\over{1+exp(e)^2}}$$.

### CDF for Logit Errors (Logit):

$$logit(e)={{exp(e)}\over{1+exp(e)}}$$.

Below I've plotted the logit and probit CDFs and PDFs:



```r
x<-seq(-5, 5, by=0.01)
plot(x, plogis(x), ylim=c(0,1), ylab="Thick Line/Logistic; Thin Line/Normal", main="CDFs")
lines(x, pnorm(x))

x<-c(-500:500)/100
plot(x, dlogis(x), ylim=c(0,1), ylab="Thick Line/Logistic; Thin Line/Normal", main="PDFs")
lines(x, dnorm(x))
```

The estimates of the model differ somewhat (why?). 

The variances of the errors are different -- $\pi^2/3$ (logit) relative to 1 (probit).

What you should note is the scale of the errors is entirely arbitrary, such that $\pi^2/3$ and 1 are not inherently meaningful. Yet, we need to make an assumption about the distribution of $y_{latent}$ to identify the model. In particular, we need to make 3 identifying assumptions (Long 1997):

(1) The mean of the error for the latent variable $e$ is 0. 

(2) $\tau$ is 0. 

(3) The variance of the error term is constant.

Consequently, if

$$p(y_{obs}=1|x)=p(y_{latent}>0|x)$$

Then,

$$p(xB+e>0|x)=p(e<-xB|x)$$

Because the distribution for $e$ must be symmetric, 

$$p(xB+e>0|x>0|x)=p(e>-xB|x)=p(e<xB|x)$$

The thing on the right is important -- it's just the CDF for the logit or probit, where 

$$p(xB+e>0|x>0|x)=p(e>-xB|x)=p(e<xB|x)=F(xB)$$

To calculate the probability of $y$, simply plug your linear function into the normal/logit CDF!

This is easiest to envision graphically. Essentially, if we identify a point $xB$, then we can find the probability of $y_{obs}=1$ to the point by using the logistic or standard normal (probit) CDFs

Yet, the actual values for $b$ will be different in the output, if we compare the logit to the probit. Recall, the logit has fatter tails. Even though the parameters differ, the the probabilties $p(y=1|x)$ will stay roughly the same. $B$ will be different insofar as the $y_{latent}$ is scaled differently. Why? 

Assue a simple model -- say the $Pr(y=1)=f_{logit}(0.2+0.25)$, and x $\in [-10,10]$


```r
x<-seq(-10,10, by=0.1)
a=.1
b=0.5

plot(x, plogis(a+b*x), ylim=c(0,1))  #Logit
plot(x, pnorm(a+b*x), ylim=c(0,1))  #Probit
```

Both models given an "s" shaped curve, but the shapes are different. The reason: The error distributions are different. There is an "indeterminacy" in the logit or probit. We can't estimate the variance -- it's not observed -- so it's fixed at something, here 1 or $\pi^2/3$. Because the errors are scaled differently, so are the parameters. Think of the OLS analogue. If I multiply the DV by a factor of 10, the slope will also change by the same amount. It's no different here. The scale of the errors -- set at something for model identification -- will change the parameter estimates. Instead, what we are likely to observe is that when we compare a logit and probit, the coefficients are different by a constant amount, about 1.6. So, 

$$1.6 \times b_{probit}=b_{logit}$$

Because the scale of the error term is entirely arbitrary, though we need to specify something, the coefficients themselves are meaningless. They cannot be directly interpreted in the way we interpret OLS coefficients. We'll see this in a bit when we discuss interpretation, but for the time being it's sufficient to simply understand that we need to make an entirely arbitrary assumption about the scale of $e$ which will cascade across the estimated coefficients in the model. ***This does not mean that the predictions from the model are arbitrary; on the contrary, the probabilities $y|x$ will be similar regardless of the assumption of whether the error is normal or logistically distributed***. Let's see with some synthetic data.


```r
x<-rnorm(5000, 0, 1) # Draw 100 values from N(0,1)
y_latent<-0.5*x
y_obs<-rep(NA, 5000)
for (i in 1:length(x)){
  y_obs[i]<-rbinom(1, 1, pnorm(y_latent[i]))
}
data<-data.frame(y_obs, x)
glm(y_obs~x, data,family=binomial("logit"))
glm(y_obs~x, data, family=binomial("probit"))
```

What you should notice is the probit coefficient is smaller than the logit coefficient -- it's also near where we expect it should be. Recall, I specified the slope coefficient at 0.5 when I generated the data (remember?). It would be strange if the data generating function (DGF), **which is known**, strongly deviates from the model estimates. Notice that if you multiple the slope by 1.6 (in the probit) you get the logit estimate

That said, things will change if we run this again -- the sample is different. If we simulate this, we're simulating a frequentism. The parameters are fixed, data are random draws from a population. Of course, we could then examine something like **unbiasedness** by doing this a bunch of times. Notice this is just unbiasedness, $E(\hat{b})=\beta$

```r

### Properties under repeated, independent samples ####
probit_fun<-function(samples){
  rep.data<-rep(NA, samples)
  for (i in 1:samples){
    x<-rnorm(5000, 0, 1) # Draw 5000 values from N(0,1)
    y_latent<-1+0.5*x
    y_obs<-rep(NA, 5000)
    for (j in 1:length(x)){
      y_obs[j]<-rbinom(1, 1, pnorm(y_latent[j]))
    }
     rep.data[i]<-coef(glm(y_obs~x, family=binomial("probit")))[2]
  }
  return(rep.data)
}


probit_fun(100)
hist(probit_fun(100))
```

## The Same Thing, The Nonlinear Regression

There is a second way to think about these models, and this approach doesn't strongly draw on a notion of an unobserved latent variable. We know the log-likelihood function for a series of Bernoulli trials.

$$  log(p(D|\theta))=\sum_{n=1}^Nlogp(y_n|\theta)=\sum_{n=1}^N[y_n log\theta+(1-y_n)log(1-\theta)] $$

In a sense, we now move the other way. Again, this much of this is in the Long reader. We have our observed data, we know that is a function of a probability, $\theta$, and we suspect  our linear predictors translate to an unobservable score that maps onto the probability. But, we can think about the problem in the reverse order -- start with the fact that we observe 0,1, we assume it's generated from $N$ bernoulli trials. $\theta$ isn't constant, it's a function of the variables. Just substitute this into the likelihood function. That's it.

First, we need, $\theta_i$. This value must be bound between 0 and 1. If we remove this restriction, then we can use that to generate a linear prediction. We want to transform $\theta$ to fall on the number line. One way is to specify:

$${{\theta_i}\over{1-\theta_i}}$$

This frees the upper bound restriction. The transformation is an "odds" that may approach $\infty$. But, we still have the zero restriction. Take the natural logarith of the odds, resulting in the log odds.

$$\eta_i=log[{{\theta_i}\over{1-\theta_i}}]$$

Such that $\eta \in [-\infty, \infty]$. 

This is the logit transformation, the log of the odds. 

Then by calculating the log-odds we have a transformed variable that we can predict with $xB$. Extend this further, by assuming $\eta_i=xB$.

$$log{{xB}\over{1-(xB)}}$$

If we calculate the inverse of this function, we may then return to the probability scale. 

$$p(y=1|x)={{exp(xB)}\over{1+exp(xB)}}$$

Which simplifies to,

$$p(y=1|x)=\theta_i={1\over{1+exp(-(xB))}}$$

Now, we have generated a function that maps the linear prediction onto the probability that $y=1$. We have ``linked'' the prediction formed by $a+bx$ onto a probability. Knowing this, we can simply input $\theta_i$ onto the likelihood function.


$$p(D|\theta)=\prod_{n=1}^N[\frac{1}{1+exp(-(xB))}]^{y_n}[1-\frac{1}{1+exp(-(xB)}]^{1-{y_n}}$$

The log of the likelihood is identical to what we've already derived. Here, we can make use of just about any **link function** that maps the linear predictors onto a [0,1] interval. In particular $\hat{xB} \rightarrow \theta_i \rightarrow y_{obs}$. You probably guessed there are two common links: The logit link and the probit link. We've explored the logit already, the probit is:

$$p(y=1|x)=\int_{-\infty}^{xB}{{1}\over{\sqrt{2 \pi}}}exp({-{t^2}\over {2}})dt=\Gamma(xB)$$. 



## A Quick Example


```r
x<-rnorm(5000, 0, 1) # Draw 100 values from N(0,1)
y_latent<-1+0.5*x
y_obs<-rep(NA, 5000)
for (i in 1:length(x)){
  y_obs[i]<-rbinom(1, 1, pnorm(y_latent[i]))
}

data<-data.frame(y_obs, x)
model<-glm(y_obs~x,  data,family=binomial("logit"))
summary(model)
probs<-plogis(predict(model, data)) ### Predicted Probabilties in the data
odds<-predict(model, data) ### Predicated log odds
```

```r
plot(density(probs), main="Probabilities")
plot(density(odds), main="Odds")

```

The log odds follow a normal distribution; the probabilities do not. ***Remember the transformation of $\theta$ into log-odds: log(pr/1-pr)?***. This is a useful interpretation of the logit. **The DV is the log odds of y. The independent variables have a linear relationship to this DV. Let's see this

```r
probs<-plogis(predict(model, data)) ### Predicted Probabilties in the data
odds<-predict(model, data) ### Predicated log odds

plot(x, odds, main="Log(Odds) and IV")
plot(x, probs, main="Probs and IV")
```

Converting between log-odds -- which again, is just logit, so it must be "log(logit-PDF)" is as simple as using $\texttt{plogis}$. If we're using a probit, the DV is no longer the log odds, so the interpretation isn't quite as direct. It's really still the same thing, but instead of $\theta$ being distributed logit, it's normal, so probit. 

It's pretty useful to generate predictions from a model. I can always pass different data into the data attribute of $\texttt{predict}$

```r
data=cbind(1,1)
print("Log Odds")
data %*% coef(model)   # Xb
print("Probabilities")
plogis(data %*% coef(model) )  # plogis(Xb)

data=cbind(1,seq(-1,1, by=0.1))
plot(data.frame(x.value=data[,2], probs=plogis(data %*% coef(model))), type="l", xlim=c(-1,1), ylim=c(0,1) ) #Vary X

```

All I've done in the plot is generate probalities at varying levels of x.
